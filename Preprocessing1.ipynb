```python
# Initialize a dictionary to store counts of spaces per column
space_counts = {}

# Loop through each column
for column in data.columns:
    # Count how many times an empty string appears in the column
    count = data[column].astype(str).str.contains("^\\s*$").sum()  # Regex to match empty strings or spaces
    space_counts[column] = count

# Convert the dictionary to a DataFrame for better readability
space_counts_data = pd.DataFrame(list(space_counts.items()), columns=['Column', 'Space Count'])

# Display the result
print(space_counts_data)
```

```python
def calculate_bmr(weight, height, age, gender):
    """
    Calculate BMR (Basal Metabolic Rate) using the Mifflin-St Jeor Equation.
    """
    if gender == 'male':
        return 10 * weight + 6.25 * height * 100 + 5 * age + 5
    else:
        return 10 * weight + 6.25 * height * 100 - 5 * age - 161

def calculate_bmi(weight, height):
    """
    Calculate BMI (Body Mass Index) using weight (kg) and height (m).
    """
    bmi = weight / (height ** 2)
    return bmi

def bmi_category(bmi):
    """
    Determine the BMI category based on the calculated BMI.
    """
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obese"

def get_caloric_needs(bmr, activity_level):
    """
    Adjust BMR based on activity level to calculate total caloric needs.
    """
    activity_multipliers = {
        'sedentary': 1.2,
        'light': 1.375,
        'moderate': 1.55,
        'active': 1.725,
        'very active': 1.9
    }
    return bmr * activity_multipliers.get(activity_level, 1.2)

def get_meal_recommendation(caloric_needs, meal_type, extracted_data):
    """
    Recommend one recipe based on caloric needs and meal type (breakfast, lunch, dinner),
    and return the recipe name, calories, and recipe category.
    """
    meal_distribution = {
        'breakfast': 0.25,
        'lunch': 0.35,
        'dinner': 0.35
    }

    meal_calories = caloric_needs * meal_distribution.get(meal_type, 0.35)
    filtered_recipes = extracted_data[extracted_data['MealType'] == meal_type].copy()

    if filtered_recipes.empty:
        return "No suitable recipe found"

    # Safely calculate the calorie difference using .loc
    filtered_recipes.loc[:, 'calorie_difference'] = (filtered_recipes['Calories'] - meal_calories).abs()
    recommended_recipe = filtered_recipes.sort_values('calorie_difference').head(1)

    return recommended_recipe[['Name', 'Calories', 'RecipeCategory']]

def recommend_all_meals(caloric_needs, extracted_data):
    """
    Recommend a breakfast, lunch, and dinner based on the calculated caloric needs.
    """
    meal_types = ['breakfast', 'lunch', 'dinner']
    recommendations = {}

    for meal_type in meal_types:
        recommendation = get_meal_recommendation(caloric_needs, meal_type, extracted_data)
        recommendations[meal_type] = recommendation

    return recommendations

# Get user input
user_weight = float(input("Enter your weight in kg: "))  # e.g., 70
user_height = float(input("Enter your height in meters: "))  # e.g., 1.75
user_age = int(input("Enter your age in years: "))  # e.g., 30
user_gender = input("Enter your gender (male/female): ").lower()  # e.g., 'male'
user_activity_level = input("Enter your activity level (sedentary, light, moderate, active, very active): ").lower()  # e.g., 'moderate'

# Calculate BMI
bmi = calculate_bmi(user_weight, user_height)
bmi_status = bmi_category(bmi)

# Display BMI and category
print(f"\nYour BMI is: {bmi:.2f}")
print(f"Your BMI category is: {bmi_status}")

# Calculate BMR and caloric needs
bmr = calculate_bmr(user_weight, user_height, user_age, user_gender)
caloric_needs = get_caloric_needs(bmr, user_activity_level)

# Display BMR and calorie needs
print(f"\nYour BMR is: {bmr:.2f} calories/day")
print(f"Your daily caloric needs (based on activity level) are: {caloric_needs:.2f} calories/day")

# Recommend three recipes for breakfast, lunch, and dinner
recommendations = recommend_all_meals(caloric_needs, extracted_data)

# Display the recommendations
print("\nRecommended Recipes:")
for meal_type, recipe in recommendations.items():
    print(f"\n{meal_type.capitalize()}:")
    print(recipe)
```
